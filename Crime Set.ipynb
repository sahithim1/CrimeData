```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Crime_Data.csv")
df.head()
```

```python
data = pd.DataFrame(df["StreetName"].value_counts())
```

```python
data[data.StreetName>500].plot(kind="bar" )
plt.show()
```

```python
data2 = pd.DataFrame(df["Offense"].value_counts())
```

```python
data2[data2.Offense>1700].plot(kind="bar" )
plt.show()
```

```python

data2[data2.Offense>1000].plot(kind='pie',subplots=True,autopct='%1.1f%%',radius = 1.23)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), shadow=True, ncol=2)
plt.ylabel("")
plt.show()
```
